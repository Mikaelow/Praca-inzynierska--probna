# Initialize data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
db = pd.read_csv("Added data.csv",sep=';')
```

```python
db #showing database
```

```python
db.hist(bins=60, figsize=(20,20))
```

# Creating histograms

```python
sns.histplot(db[['A_P1', 'A_P2', 'B', 'C', 'D', 'E', 'F']], bins=1000)
plt.xlim(-10, 10)
```

## Choosing most differend data to `A_P1`

```python
sns.histplot(db[['A_P1','B','F']])
plt.xlim(-10, 10)
```

 ## Creating boxplots

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 8))  # 2 wiersze, 1 kolumna
#boxplot with showfliers on 
sns.boxplot(ax=axs[0], data=db[['A_P1', 'A_P2', 'B', 'C', 'D', 'E', 'F']])
axs[0].set_title("Boxplot with showfliers on")

#boxplot without showfliers 
sns.boxplot(ax=axs[1], data=db[['A_P1', 'A_P2', 'B', 'C', 'D', 'E', 'F']],showfliers=False)
axs[1].set_title("Boxplot without showfliers on")

```

```python

#boxplot without showfliers and F bearing
sns.boxplot(data=db[['A_P1', 'A_P2', 'B', 'C', 'D', 'E']],showfliers=False)
plt.title("Boxplot without showfliers on and F bearing")
```
